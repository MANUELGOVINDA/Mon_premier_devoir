```python
# moyenne et ecart_type

values = slice(3000,7000)
N = len(values)

def moyenne (values):
    return sum(values)/N

def ecart_type(values):
    v = 0
    M = moyenne(values)
    for i in range (N):
        v = v+(values[i]-M)**2
    return math.sqrt((v/(N-1)))
```
